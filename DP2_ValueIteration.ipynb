```python
import numpy as np

#状態価値関数をランダムに初期化
V=np.random.rand(4,1)*10
print('V')
print(V)

#状態遷移行列
Pas=np.array([[0.9, 0.1,0,0],[0.1,0.9,0,0],[0.9,0.1,0,0],[0,0.1,0.9,0],
            [0,0.9,0.1,0],[0,0,0.1,0.9],[0,0,0.9,0.1],[0,0,0.1,0.9]])
print('Pas')
print(Pas)

#報酬ベクトル (状態がS4の時のみ報酬)
#推移確率行列にRを書けた数値が報酬の期待値
Rs=np.array([0,0,0,1])

#割引率
gamma=0.9

theta=0.000000001
```

```python
t=1
while True:
    print('time step ' + str(t))
    delta=0
    Vnext=Rs.reshape(4,1)+gamma*V #4次元の列ベクトル
    Q=np.dot(Pas,Vnext).reshape(4,2) #状態・行動価値関数
    print('Q')
    print(Q)
    Vupdate=Q.max(1).reshape(4,1) #各状態で最適行動を選ぶときの状態価値関数
    delta=max(np.absolute(Vupdate-V))
    V=Vupdate
    print('V')
    print(V)
    t+=1
    if delta<theta or t>10000:
        break
        
print('Optimal policy')
print(Q.argmax(1))
```
